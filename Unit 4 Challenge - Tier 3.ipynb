#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# Transpose and reset the index of the properties dataframe
properties_T = properties.T.reset_index()

# Check the data
properties_T.head()
```

```python
# Assign the first row to be the column headers
properties_T.columns = properties_T.iloc[0]

# Drop the first row
properties_T = properties_T.drop(0)

#Check the data
properties_T.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename 'Unnamed: 0' to 'London_Borough' and NaT to 'ID'
properties_T = properties_T.rename(columns = {'Unnamed: 0':'London_Borough', pd.NaT: 'ID'})

# Check the data
properties_T.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt columns into rows
properties_melted = pd.melt(properties_T, id_vars = ['London_Borough', 'ID'])

# Check the data
properties_melted.info()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Rename the last 2 columns to 'Month' and 'Average_price'
properties_melted = properties_melted.rename(columns = {0: 'Month', 'value': 'Average_price'})

# Change the data type of 'Average_price' to float
properties_melted['Average_price'] = pd.to_numeric(properties_melted['Average_price'])

# Change the data type of 'Month' to datetime
properties_melted['Month'] = pd.to_datetime(properties_melted['Month'])

# Check the data
properties_melted.info()
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Filter out nonboroughs
boroughs_df = properties_melted[properties_melted['London_Borough'].isin(borough_list)]

# Check the data
boroughs_df.info()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Create a variable that filters data only for Brent and assign to Brent_prices
Brent_prices = boroughs_df[boroughs_df['London_Borough'] == 'Brent']

# Create a line plot of Brent prices and assign to ax
ax = Brent_prices.plot(kind = 'line', x = 'Month', y = 'Average_price')

ax.set_ylabel('Price')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column.

```python
# Create a copy of boroughs_df and assign it to boroughs_df_by_year 
boroughs_df_by_year = boroughs_df.copy()

# # Extract the year from the 'Month' column
boroughs_df_by_year.loc[:, 'Year'] = boroughs_df_by_year['Month'].dt.year

# Check the data
boroughs_df_by_year.info()
```

```python
# Create a new DataFrame boroughs_dfg that groups the data by London_Borough and Year and resets the index
boroughs_dfg = boroughs_df_by_year.groupby(['London_Borough', 'Year']).agg({'Average_price': 'mean'}).reset_index()

# Check the data
boroughs_dfg.head()
```

```python
# Compute the percentage change in Average_price from 1998 to 2018
boroughs_dfg_1998 = boroughs_dfg[boroughs_dfg['Year'] == 1998]
boroughs_dfg_2018 = boroughs_dfg[boroughs_dfg['Year'] == 2018]
boroughs_dfg_1998_2018 = boroughs_dfg_1998.merge(boroughs_dfg_2018, on = 'London_Borough', suffixes = ('_1998', '_2018'))


boroughs_dfg_1998_2018
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(d):
    y1998 = float(d.loc[d['Year']==1998, 'Average_price'].iloc[0]) 
    y2018 = float(d.loc[d['Year']==2018, 'Average_price'].iloc[0])  
    ratio = [y1998 / y2018]
    return ratio

# Create empty dictionary to store ratios
final = {}

# Loop through each borough and calculate the ratio
for b in boroughs_dfg['London_Borough'].unique():
    borough = boroughs_dfg[boroughs_dfg['London_Borough'] == b]
    final[b] = create_price_ratio(borough)
print(final)

# Make a variable called boroughs_ratios, and assign it the result of calling the DataFrame method on the dictionary final. 
boroughs_ratios = pd.DataFrame(final)

# Check the data
boroughs_ratios.head()
```

```python
# Transpose the data and reset the index
boroughs_ratios_T = boroughs_ratios.T.reset_index()

# Check the data
boroughs_ratios_T.head()
```

```python
# Rename the columns 'London_Borough' and 'Ratio'
boroughs_ratios = boroughs_ratios_T.rename(columns = {'index':'London_Borough', 0: 'Ratio'})

# Check the data
boroughs_ratios.head()
```

```python
# Sort the data in descending order
boroughs_ratios_sorted = boroughs_ratios.sort_values(by = 'Ratio', ascending = False)

# Check the data
boroughs_ratios_sorted
```

```python
# Plot the boroughs_ratios_sorted data to see the greatest price changes
ax = boroughs_ratios_sorted.plot(kind = 'bar', x = 'London_Borough', y = 'Ratio')

# Set the tick labels for each borough
ax.set_xticklabels(boroughs_ratios_sorted.London_Borough)
```

```python
# Recreate the function but to visualize the percent change in prices from 1998 to 2018
def create_percent_change(d):
    y1998 = float(d.loc[d['Year']==1998, 'Average_price'].iloc[0]) 
    y2018 = float(d.loc[d['Year']==2018, 'Average_price'].iloc[0])  
    percent_change = [((y2018 - y1998) / y1998) * 100]
    return percent_change

# Create empty dictionary to store percent changes
final = {}

# Loop through each borough and calculate the percent change
for b in boroughs_dfg['London_Borough'].unique():
    borough = boroughs_dfg[boroughs_dfg['London_Borough'] == b]
    final[b] = create_percent_change(borough)
print(final)

# Make a variable called boroughs_percent_change, and assign it the result of calling the DataFrame method on the dictionary final. 
boroughs_percent_change = pd.DataFrame(final)

# Transpose the data and reset the index
boroughs_percent_change_T = boroughs_percent_change.T.reset_index()

# Rename the columns 'London_Borough' and 'Percent_change'
boroughs_percent_change = boroughs_percent_change_T.rename(columns = {'index':'London_Borough', 0: 'Percent_change'})

# Sort the data in descending order
boroughs_percent_change_sorted = boroughs_percent_change.sort_values(by = 'Percent_change', ascending = False)

# Plot the boroughs_percent_change_sorted data to see the greatest price changes
ax = boroughs_percent_change_sorted.plot(kind = 'bar', x = 'London_Borough', y = 'Percent_change')

# Set the tick labels for each borough
ax.set_xticklabels(boroughs_percent_change_sorted.London_Borough)
```
